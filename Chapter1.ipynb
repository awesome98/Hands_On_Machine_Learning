# 1.1 머신러닝이란?

* 데이터에서부터 학습하도록 컴퓨터를 프로그래밍하는 과학이다.
* 명시적인 프로그래밍 없이 컴퓨터가 학습하는 능력을 갖추게 하는 연구 분야다.
* 어떤 작업 T에 대한 컴퓨터 프로그램의 성능을 P로 측정했을 때 경험 E로 인해 성능이 향상됐다면,
  이 컴퓨터 프로그램은 작업 T와 성능 측정 P에 대해 경험 E로 학습한 것이다.
* 시스템이 학습하는 데 사용하는 샘플을 training set라고 하고 각 훈련 데이터를 훈련 사례(혹은 샘플)라고 합니다.
  이 경우 작업 T는 새로운 메일이 스팸인지 구분하는 것이고, 경험 E는 training data이며, 성능 측정 P는 직접 정의해야 함.
  예를 들어 정확히 분류된 메일의 비율을 P로 사용할 수 있으며, 이 성능 측정을 정확도(accuracy)라고 부르며 분류 작업에 자주 사용됨.

# 1.2 왜 머신러닝을 사용하는가?

* 가끔 예상치 못한 연관관계나 새로운 추세가 발견되기도 해서 해당 문제를 더 잘 이해하도록 도와준다.
* 머신러닝 기술을 적용해서 대용량의 데이터를 분석하면 겉으로는 보이지 않던 패턴을 발결할 수 있는데 이를 **데이터 마이닝**이라 한다.

# 1.4 머신러닝 시스템의 종류

## 1.4.1 지도 학습과 비지도 학습

* 머신러닝 시스템을 '학습하는 동안의 감독 형태나 정보량'에 따라 분류할 수 있다.
    * 지도 학습, 비지도 학습, 준지도 학습, 강화 학습

### 지도 학습

* 알고리즘에 주입하는 훈련 데이터에 **레이블**이라는 원하는 답이 포함됨.
* **분류**가 전형적인 지도 학습 작업
* **예측변수**라 부르는 **특성**을 사용해 중고차 가격 같은 **타겟**수치를 예측하는 작업. 이런 종류의 작업을 **회귀**라고 부름.
    * Note 머신러닝에서 레이블의 범주를 클래스(class)라고 부름. 속성은 데이터 타입(컬럼)을 말하고,
      특성은 일반적으로 속성과 값이 합쳐진 것을 의미. ex) 주행거리 = 15000
* 지도학습 알고리즘들
    1. k-NN 알고리즘
    2. 선형 회귀
    3. 로지스틱 회귀
    4. SVM(서포트 벡터 머신)
    5. 결정 트리와 랜덤 포레스트
    6. NN(신경망)

### 비지도 학습

* 비지도 학습은 훈련 데이터에 레이블이 없다. 시스템이 아무런 도움 없이 학습함.
* 비지도 학습 알고리즘
    * 군집
        - k-means
        - DBSCAN
        - 계층 군집 분석(HCA)
        - 이상치 탐지와 특이치 탑지
        - 원-클래스
        - 아이솔레이션 포레스트
    * 시각화와 차원 축소
        - PCA(주성분 분석)
        - 커널 PCA
        - 지역적 선형 임베딩
        - t-SNE
    * 연관 규칙 학습
        - Apriori
        - Eclat

### 준지도 학습

* 일부만 레이블이 있는 데이터를 다루는 알고리즘
* 지도 학습과 비지도 학습의 조합
* 심층 신뢰 신경망(DBN)은 여러 겹으로 쌓은 제한된 볼츠만 머신(RBN)이라 불리는 비지도 학습에 기초

### 강화 학습

* 학습하는 시스템을 **에이전트**라고 부르며 환경을 관찰해서 행동을 실행하고 그 결과로 **보상**(또는 **벌점**)을 받는다.
* 시간이 지나면서 가장 큰 보상을 얻기 위해 **정책**이라고 부르는 최상의 전략을 스스로 학습함
* **정책**은 주어진 상황에서 에이전트가 어떤 행동을 선택할지 정의함.

## 1.4.2 배치 학습과 온라인 학습

### 배치 학습

* 가용한 데이터를 모두 사용해 훈련시켜야 함. 일반적으로 이 방식은 시간과 자원을 많이 소모하여 오프라인에서 수행
* 학습한 것을 단지 적용만 한다. 이를 **오프라인 학습**이라고 한다.
* 데이터 양이 많으면 배치 학습 알고리즘을 사용하는데 불가능할 수도 있다.

### 온라인 학습(점진적 학습)

* 데이터를 순차적으로 한 개씩 또는 **미니배치**라 부르는 작은 묶음 단위로 주입하여 시스템을 훈련시킨다.
* 매 학습 단계가 빠르고 비용이 적게 들어서 시스템은 데이터가 도착하는 대로 즉시 학습 가능
* 연속적으로 데이터를 받고 빠른 변화에 스스로 적응해야하는 시스템에 적합함.
* 중요한 파라미터 하나는 변화하는 데이터에 얼마나 빠르게 적응할 것인지를 비율로 나타낸 것이 **학습률**이다.
* 학습률을 높이면 시스템이 데이터에 빠르게 적응하지만, 예전 데이터를 금방 잊어버린다. 반대로 학습률이 낮으면 시스템 관성이 커져서 더 느리게 학습된다. 하지만 새로운 데이터에 있는 잡음, 대표성 없는 데이터 포인트에 덜 민감하다.

## 1.4.3 사례 기반 학습과 모델 기반 학습

* 어떻게 **일반화**되는가에 따라 분류
* NOTE : 훈련데이터에서 높은 성능을 내는 것이 좋지만 진짜 목표는 새로운 샘플에 잘 작동하는 모델!

### 사례 기반 학습

* 시스템이 훈련 샘플을 기억함으로써 학습하고 유사도 측정을 사용해 새로운 데이터와 학습한 샘플을 비교하는 식을 일반화 함.

### 모델 기반 학습
* 샘플로부터 일반화시키는 다른 방법은 샘플들의 모델을 만들어 **예측**에 사용하는 것으로 모델 기반 학습이라고 한다.

# 1.5 머신러닝의 주요 도전 과제

## 1.5.1 충분하지 않은 양의 훈련 데이터
* 데이터 많으면 모델 학습에 무조건 좋음

## 1.5.2 대표성 없는 훈련 데이터
* 샘플링 잡음(샘플이 작을 때 우연에 의한 대표성 없는 데이터), 샘플링 편향(큰 샘플도 표본추출방법이 잘못되면 대표성X)

## 1.5.3 낮은 품질의 데이터
* 훈련 데이터에 에러, 이상치, 잡음으로 가득하면 패턴 찾기 어려우므로 훈련 데이터 정제가 필요함.

## 1.5.4 관련 없는 특성
* 성공적인 머신러닝 프로젝트의 핵심 요소는 훈련에 사용할 좋은 특성들을 찾는 것

## 1.5.5 훈련 데이터 과대적합
* 과대적합(overfitting) : 모델이 훈련 데이터에 너무 잘 맞지만 일반성이 떨어진다
* 과대적합 해결방법
    * 파라미터 수가 적은 모델 선택, 훈련 데이터안의 특성 수 줄이거나, 모델에 제약을 가하여 단순화
    * 훈련 데이터를 더 많이 모은다.
    * 훈련 데이터를 전처리하여 잡음을 줄인다.
* 하이퍼파리미터는 학습 알고리즘으 파라미터, 학습 알고리즘에 영향을 받지 않으며, 훈련 전에 미리 지정, 훈련 동안 상수로 고정됨

## 1.5.6 훈련 데이터 과소적합
* 모델이 너무 단순해서 데이터의 내재된 구조를 학습하지 못할 때 발생
* 과소적합 해결방법
    * 모델 파라미터가 더 많은 강력한 모델을 선택
    * 학습 알고리즘에 더 좋은 특성을 제공
    * 모델에 제약을 줄임(규제 하이퍼파라미터를 감소시킴)

# 1.6 테스트와 검증
* 모델이 새로운 샘플에 얼마나 잘 일반화될지 아는 유일한 방법은 새로운 샘플(test set)에 실제로 적용해보는 것
* **훈련 데이터를 training set와 test set 두개로 나누는 것**
* 훈련 오차가 낮지만 일반화 오차가 높다면 이는 모델이 훈련 데이터에 과대적합되었다는 뜻
* TIP : 보통 80%는 훈련용, 20%는 테스트용으로 하지만 데이터셋 크기에 따라 다르다. 샘플이 천만 개라면 1%만 떼도 테스트 셋으로 10만개가 존재하고 일반화 오차를 추정하는데 충분한 크기일 것이다.

## 1.6.1 하이퍼파라미터 튜닝과 모델 선택
* 홀드아웃 검증 : 훈련 셋의 일부를 떼어내어 여러 후보 모델을 평가하고 베스트 원을 선택. 새로운 홀드아웃 셋을 검증세트라 부름
    1. 검증 셋을 제외한 훈련 세트에서 다양한 하이퍼파라미터 값을 가진 여러 모델을 훈련한다.
    2. 검증 셋에서 가장 높은 성능을 내는 모델을 선택
    3. 홀드아웃 검증이 끝나면 선택된 최선의 모델을 전체 훈련셋(검증셋포함)에서 다시 훈련하여 최종 모델을 만든다
    4. 마지막으로 최종 모델을 테스트셋에서 평가하여 일반화 오차를 추정한다.
* 교차검증 : 작은 검증셋을 여러 개(서로 다른) 사용해 반복적으로 검증하는 것
    1. 검증 셋마다 나머지 데이터에서 훈련한 모델을 해당 검증 셋에서 평가
    2. 모든 모델의 평가를 평균하면 훨씬 정확한 성능을 측정할 수 있음(검증셋의 개수에 비례해 훈련 시간이 늘어남)
